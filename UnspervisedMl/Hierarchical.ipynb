Algorithm Overview: Hierarchical clustering builds a tree-like structure (dendrogram) by iteratively merging or splitting clusters based on a linkage criterion, forming a hierarchy of clusters.
\

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.metrics import silhouette_score
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
x,y=make_blobs(n_samples=100,n_features=2,centers=3)
```

```python
y
```

```python
# creating the dendogram to choose the best size of the cluster
import scipy.cluster.hierarchy as shc
```

```python
dend=shc.dendrogram(shc.linkage(x,method='ward'))
```

```python
model= AgglomerativeClustering(n_clusters=3,linkage='ward')
```

```python
acmodel=model.fit_predict(x)
```

```python
color = ['blue', 'purple', 'red', 'green', 'orange', 'cyan', 'magenta', 'brown', 'violet', 'gold']
for i in range(0,3):
  plt.scatter(x=x[acmodel==i,0],y=x[acmodel==i,1],color=color[i])
```

```python
score=silhouette_score(x,acmodel)
```

```python
print(score)
```
